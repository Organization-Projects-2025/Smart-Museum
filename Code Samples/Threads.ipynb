```python
# Function to print numbers from 1 to 28
def print_numbers():
    for i in range(1, 27):
        print(i)


# Function to print letters from A to Z
def print_letters():
    for letter in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":
        print(letter)
```

```python
print_letters()
print_numbers()
```

```python
import threading

# Create two threads
thread1 = threading.Thread(target=print_numbers)
thread2 = threading.Thread(target=print_letters)

# Start the threads
thread1.start()
thread2.start()

# Wait for both threads to finish
thread1.join()
thread2.join()


# Print a message when both threads are done
print("Both threads have finished.")
```

# Using Locks

```python
import threading

# Lock for synchronization
lock = threading.Lock()


# Function to print numbers from 1 to 28
def print_numbers():
    for i in range(1, 29):
        with lock:
            print(i)


# Function to print letters from A to Z
def print_letters():
    for letter in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":
        with lock:
            print(letter)


# Create two threads
thread1 = threading.Thread(target=print_numbers)
thread2 = threading.Thread(target=print_letters)

# Start the threads
thread1.start()
thread2.start()

# Wait for both threads to finish
thread1.join()
thread2.join()

# Print a message when both threads are done
print("Both threads have finished.")
```
