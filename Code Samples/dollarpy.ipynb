```python
import mediapipe as mp  # Import mediapipe
import cv2  # Import opencv
from dollarpy import Recognizer, Template, Point

mp_drawing = mp.solutions.drawing_utils  # Drawing helpers
mp_drawing_styles = mp.solutions.drawing_styles

mp_holistic = mp.solutions.holistic  # Mediapipe Solutions
```

```python
templates = []  # list of templates for $1 training
```

```python
def getPoints(videoURL, label):
    cap = cv2.VideoCapture(videoURL)  # web cam =0 , else enter filename
    # Initiate holistic model
    with mp_holistic.Holistic(
        min_detection_confidence=0.5, min_tracking_confidence=0.5
    ) as holistic:
        points = []
        left_shoulder = []
        right_shoulder = []
        left_elbows = []
        right_elbows = []
        left_wrist = []
        right_wrist = []
        left_pinky = []
        right_pinky = []
        left_index = []
        right_index = []
        left_hip = []
        right_hip = []
        while cap.isOpened():
            ret, frame = cap.read()

            # Recolor Feed
            if ret == True:

                image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                image.flags.writeable = False

                # Make Detections
                results = holistic.process(image)
                # print(results.face_landmarks)

                # Recolor image back to BGR for rendering
                image.flags.writeable = True
                image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

                # Drawing on Frame (You can remove it)
                # 2. Right hand
                mp_drawing.draw_landmarks(
                    image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS
                )

                # 3. Left Hand
                mp_drawing.draw_landmarks(
                    image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS
                )

                # 4. Pose Detections
                mp_drawing.draw_landmarks(
                    image,
                    results.pose_landmarks,
                    mp_holistic.POSE_CONNECTIONS,
                    landmark_drawing_spec=mp_drawing_styles.get_default_pose_landmarks_style(),
                )
                # Export coordinates

                try:

                    # add points of wrist , elbow and shoulder
                    left_shoulder.append(
                        Point(
                            results.pose_landmarks.landmark[11].x,
                            results.pose_landmarks.landmark[11].y,
                            1,
                        )
                    )
                    right_shoulder.append(
                        Point(
                            results.pose_landmarks.landmark[12].x,
                            results.pose_landmarks.landmark[12].y,
                            2,
                        )
                    )
                    left_elbows.append(
                        Point(
                            results.pose_landmarks.landmark[13].x,
                            results.pose_landmarks.landmark[13].y,
                            3,
                        )
                    )
                    right_elbows.append(
                        Point(
                            results.pose_landmarks.landmark[14].x,
                            results.pose_landmarks.landmark[14].y,
                            4,
                        )
                    )
                    left_wrist.append(
                        Point(
                            results.pose_landmarks.landmark[15].x,
                            results.pose_landmarks.landmark[15].y,
                            5,
                        )
                    )
                    right_wrist.append(
                        Point(
                            results.pose_landmarks.landmark[16].x,
                            results.pose_landmarks.landmark[16].y,
                            6,
                        )
                    )
                    left_pinky.append(
                        Point(
                            results.pose_landmarks.landmark[17].x,
                            results.pose_landmarks.landmark[17].y,
                            7,
                        )
                    )
                    right_pinky.append(
                        Point(
                            results.pose_landmarks.landmark[18].x,
                            results.pose_landmarks.landmark[18].y,
                            8,
                        )
                    )
                    left_index.append(
                        Point(
                            results.pose_landmarks.landmark[19].x,
                            results.pose_landmarks.landmark[19].y,
                            9,
                        )
                    )
                    right_index.append(
                        Point(
                            results.pose_landmarks.landmark[20].x,
                            results.pose_landmarks.landmark[20].y,
                            10,
                        )
                    )
                    left_hip.append(
                        Point(
                            results.pose_landmarks.landmark[23].x,
                            results.pose_landmarks.landmark[23].y,
                            11,
                        )
                    )
                    right_hip.append(
                        Point(
                            results.pose_landmarks.landmark[24].x,
                            results.pose_landmarks.landmark[24].y,
                            12,
                        )
                    )

                except:
                    pass

                cv2.imshow(label, image)
            else:
                cap.release()
                cv2.destroyAllWindows()
                cv2.waitKey(100)
                break

            if cv2.waitKey(10) & 0xFF == ord("q"):
                cap.release()
                cv2.destroyAllWindows()
                cv2.waitKey(100)
                break

    cap.release()
    cv2.destroyAllWindows()
    points = (
        left_shoulder
        + right_shoulder
        + left_elbows
        + right_elbows
        + left_wrist
        + right_wrist
        + left_pinky
        + right_pinky
        + left_index
        + right_index
        + left_hip
        + right_hip
    )
    print(label)
    return points
```

```python
# hook correct
vid = "Dataset/hook/hook correct.mp4"
points = getPoints(vid, "hook correct")
temp_2 = Template("hook correct", points)
templates.append(temp_2)

# hook no defense no torso twist
# vid = "Dataset\hook\hook wrong no torso twist no defense.mp4"
# points = getPoints(vid,"hook wrong no torso twist no defense")
# temp_2 = Template('hook wrong no torso twist no defense', points)
# templates.append(temp_2)

# hook wrong no torso twist
vid = "Dataset/hook/hook wrong no torso twist.mp4"
points = getPoints(vid, "hook wrong no torso twist")
temp_2 = Template("hook wrong no torso twist", points)
templates.append(temp_2)

# hook wrong no defense
vid = "Dataset/hook/hook wrong no torso twist no defense.mp4"
points = getPoints(vid, "hook wrong no defense")
temp_2 = Template("hook wrong no defense", points)
templates.append(temp_2)

# UPPER
# uppercut correct
vid = "Dataset/uppercut/uppercut correct.mp4"
points = getPoints(vid, "uppercut correct")
temp_2 = Template("uppercut correct", points)
templates.append(temp_2)

# uppercut wrong (no torso & ankle  twist)
vid = "Dataset/uppercut/uppercut wrong (no torso & ankle  twist).mp4"
points = getPoints(vid, "uppercut wrong (no torso & ankle  twist)")
temp_2 = Template("uppercut wrong (no torso & ankle  twist)", points)
templates.append(temp_2)

# uppercut wrong no defense
vid = "Dataset/uppercut/uppercut wrong no defense.mp4"
points = getPoints(vid, "uppercut wrong no defense")
temp_2 = Template("uppercut wrong no defense", points)
templates.append(temp_2)

# right stroke wrong no defense
vid = "Dataset/right/(right) correct.mp4"
points = getPoints(vid, "(right) correct")
temp_2 = Template("(right) correct", points)
templates.append(temp_2)

# right stroke wrong hand down no torso twist
vid = "Dataset/right/(right) hand down no torso twist.mp4"
points = getPoints(vid, "(right) hand down no torso twist")
temp_2 = Template("(right) hand down no torso twist", points)
templates.append(temp_2)

# right wrong  wrong return
vid = "Dataset/right/(right) wrong return.mp4"
points = getPoints(vid, "(right) wrong return")
temp_2 = Template("(right) wrong return", points)
templates.append(temp_2)
```
