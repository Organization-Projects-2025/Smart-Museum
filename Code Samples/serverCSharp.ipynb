```python
import socket

mySocket = socket.socket()


def connect_socket():
    global conn, addr
    mySocket.bind(("localhost", 5000))
    mySocket.listen(5)
    conn, addr = mySocket.accept()
    print("Device Connected")


connect_socket()

msg = bytes("Hello from the Server", "utf-8")
conn.send(msg)
msg = bytes("q", "utf-8")
conn.send(msg)
```

```python
# This is used to close the socket connection after finishing the communication
mySocket.close()
```
